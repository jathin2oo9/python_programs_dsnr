````python
# Program to print numbers between 1 to 20 using while loop
num = 1
while num <= 20:
    print(num)
    num += 1
````

````python
# Program to find the sum of all even numbers between 1 and 100 using while loop
num = 1
sum_even = 0
while num <= 100:
    if num % 2 == 0:
        sum_even += num
    num += 1
print("The sum of all even numbers between 1 and 100 is:", sum_even)
````

````python
# Program to find the sum of all odd numbers
# and product of all even numbers between 1 and 100 using while loop
num = 1
sum_odd = 0
product_even = 1
while num <= 100:
    if num % 2 == 0:           # even number
        product_even *= num
    else:                      # odd number
        sum_odd += num
    num += 1
print("Sum of all odd numbers between 1 and 100 is:", sum_odd)
print("Product of all even numbers between 1 and 100 is:", product_even)
````

````python
# Program to find the product of all odd numbers between 10 and 30 using while loop

num = 10
product = 1

while num <= 30:
    if num % 2 != 0:      # check if the number is odd
        product *= num     # multiply it with the product
    num += 1               # increase number by 1

print("The product of all odd numbers between 10 and 30 is:", product)

````
