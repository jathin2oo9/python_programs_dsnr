````python
# write a program to find the given number i positive or negative or zero 
num=int(input("please enter a number:"))
if num > 0:
    print("The number is Positive.")
elif num < 0:
    print("The number is Negative.")
else:
    print("The number is Zero.")
````

````python
'''
# write a program to find the grades based on student marks 
1. if marks in b/w 80 to 100 grade is A 
2. if marks in b/w 60 to 79 grade is B 
3. if marks in b/w 40 to 59 grade is C
4. if marks in b/w 25 to 39 grade is D 
5. if marks in b/w 0 to 24 grade fail 
6. if mark other than above output shoub be please check valid marks
'''
marks=int(input("please emter studennt marks here:"))
if marks >= 80 and marks <= 100:
    print("Grade: A")
elif marks >= 60 and marks <= 79:
    print("Grade: B")
elif marks >= 40 and marks <= 59:
    print("Grade: C")
elif marks >= 25 and marks <= 39:
    print("Grade: D")
elif marks >= 0 and marks <= 24:
    print("Fail")
else:
    print("Please check valid marks.")
````
